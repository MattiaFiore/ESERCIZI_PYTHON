# Pokemon di tipo fuoco


```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

### Opening the dataset

```python
df = pd.read_csv("/home/mattiafiore/Desktop/Esercizi_Python/Pandas/dataset/pokemon.csv")
# It is possible to show the first 5 rows
df.head(5)
total_pokemons = df.shape[0]
```

### Selecting only Fire pokemon 

```python
df2 = df[(df['Type 1'] == 'Fire') | (df['Type 2'] == 'Fire')]
df2
```

```python
total_fire_pokemons = df2.shape[0]
```

```python
print(f"Percentage of fire pokemon {total_fire_pokemons/total_pokemons*100}%")
```

### I pokemon fuoco hanno una statistica di attacco piu alta? 

```python
attacco_medio = df["Attack"].mean()
std_attacco = df["Attack"].std()
attacco_medio_fuoco = df2["Attack"].mean()
std_attacco_fuoco = df2["Attack"].std()
```

```python
print(f"Attacco medio {attacco_medio} std: {std_attacco}")
```

```python
print(f"Attacco medio pokemon fuoco: {attacco_medio_fuoco} std: {std_attacco_fuoco}")
```

### Let's make a graph 

```python
plt.hist(df['Attack'], bins=12, alpha = 0.5, density=True)
plt.hist(df2["Attack"], bins=12, alpha=0.5, density=True)

# Diamo nomi agli assi e al grafico 
plt.xlabel("Attack")
plt.ylabel("Density")
plt.title("Distribuizione attacco")
plt.legend(["Tutti pokemon", "Tipo fuoco"])
```

### Piu grafici
Fare questo grafico su tutte le statistiche sarebbe molto noioso per questo motivo conviene creare uno script che salva le foto su una cartella per tutte le statistiche.

```python
attributi = ['HP', 'Attack', 'Defense', 'Sp. Atk',
       'Sp. Def', 'Speed']

for attributo in attributi: 
    plt.figure()
    plt.hist(df[attributo], bins=10, alpha = 0.5, density=True)
    plt.hist(df2[attributo], bins=10, alpha=0.5, density=True)

    # Diamo nomi agli assi e al grafico 
    plt.xlabel(attributo)
    plt.ylabel("Density")
    plt.title(f"Distribuizione {attributo}")
    plt.legend(["Tutti pokemon", "Tipo fuoco"])
    plt.savefig(f'{attributo}_graph.png')
```
